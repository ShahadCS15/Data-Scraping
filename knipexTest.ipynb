```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import requests
from requests import exceptions
from bs4 import BeautifulSoup
import pandas as pd

import urllib3
```

```python
urllib3.disable_warnings()
```

```python
def page_driver(item_category, item_id):
    try:
        options = webdriver.ChromeOptions()
        options.headless = True
        driver = webdriver.Chrome('C:\\Users\\Shahad\\Downloads\\chromedriver_win32\\chromedriver.exe', options=options)  
        driver.get('https://www.knipex.com/nc/en/home/')
        #driver.implicitly_wait(10)

        search_bar = driver.find_element_by_id('s')
        search_bar.send_keys(item_id)

        submit_button = driver.find_element_by_id('go')
        submit_button.click()

        group_item = driver.find_element_by_link_text(item_category)
        group_item.click()

        select_item = driver.find_element_by_link_text(item_id)
        select_item.click()

        url = driver.current_url
        return url
    except:
        pass
```

```python
df = pd.read_excel(r'Knipex Dataset.xlsx')
```

```python
def getURLs():
    url_lst=[]
    for i in df.index:
        url = page_driver(df['Sub-Category'][i], df['Identifier'][i])
        url_lst.append(url)
    return url_lst
```

```python
urls = getURLs()
#print(urls)
df['webpage_to_be_scraped'] = urls
```

```python
def create_soup(url):
    try:
        r = requests.get(url, verify=False)
        soup = BeautifulSoup(r.text, 'lxml')
        #print(soup)
        return soup
    except:
        return 'N/A'
```

```python
soup = create_soup('https://www.knipex.com/index.php?id=1216&L=1&page=art_detail&parentID=1363&groupID=1392&artID=1001')
```

```python
def get_Specs(soup):
    try:
        inner_dictionary = {}
        
        div = soup.find('div', {'id' : 'fragment-1'})
        body = div.contents[0].find_all('tr')
        for row in body:
            cols = row.findChildren(recursive=False)
            cols = [ele.text.strip() for ele in cols]
            cols.remove(cols[0])
            inner_dictionary.update({cols[0] : cols[1]})
            print(inner_dictionary)
        return inner_dictionary
    
    except:
        pass
        # try return 'N/A'
```

```python
def get_img(soup):
    try:

        img = soup.find('a', {'class' : 'lightbox zoom'}).get('href')
        img = 'https://www.knipex.com'+ img
        print(img)
        return img
    except:
        pass
```
