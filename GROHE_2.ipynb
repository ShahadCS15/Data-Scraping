```python
def page_driver(item_id):
    try:

        options = webdriver.ChromeOptions()
        options.headless = True
        driver = webdriver.Chrome('C:\\Users\\Shahad\\Downloads\\chromedriver_win32\\chromedriver.exe', options=options)  
        #driver = webdriver.Chrome('C:\\Users\\Shahad\\Downloads\\chromedriver_win32\\chromedriver.exe')  
        #driver.maximize_window()
        driver.get('https://www.grohe-mena.com/en_cy/')
        
        #Main page1
        search_icon = driver.find_element_by_id('flyout-search')
        search_icon.click()
        

        #Search bar page2
        search_bar = driver.find_element_by_xpath("//input[@class='autosuggest__input form-control form-control-lg']")
        search_bar.send_keys(item_id)

        search_btn = driver.find_element_by_class_name('autosuggest__searchBtn')
        search_btn.click()
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight/3.0)")
        
        #products list page3
        driver.implicitly_wait(10)
        select_item = driver.find_element_by_class_name('teaser-product-row__image')
        select_item.click()
        
        #value in new page
        new_value = driver.find_element_by_xpath("//div[@class='product-box__tableCell product-box__tableCell--label']")
        url = driver.current_url
        return url
    
    except:
        pass
```

```python
def getURLs():
    url_lst=[]
    for i in df.index:
        url = page_driver(str(df['Item Code'][i]))
        url_lst.append(url)
    return url_lst
```

```python
def create_soup(url):
    try:
        r = requests.get(url, verify=False)
        soup = BeautifulSoup(r.text)
        return soup
    except:
        pass
```

```python
def get_prod_info(soup):
    try:
        title_lst = []
        titles = soup.findAll('div', {'class' : 'product-box__tableCell product-box__tableCell--label'})
        for title in titles:
            title_lst.append(title.text)
            title.find_next()
            
        values_lst = []
        values = soup.findAll('div', {'class' : 'product-box__tableCell product-box__tableCell--value'})
        for val in values:
            values_lst.append(val.text.strip())
            val.find_next()
            
        info_dic = {}
        for i in range(len(title_lst)):
            info_dic.update({ title_lst[i] : values_lst[i] })
        return info_dic
    except:
        pass
```

```python
def get_img(url):
    try:
        img_dic = {}
        num = 1
        driver.get(url)
        images = driver.find_elements_by_xpath('//a[@class="image__link image__link--enabled"]')
        for img in images:
            img_link = img.get_attribute('href')
            img_dic.update({ 'image_'+str(num) : img_link })
            num = num + 1
        return img_dic
    except:
        pass
```

```python
def get_tech_and_CAD_files(soup):
    try:
        files_dic = {}
            
        links = soup.findAll('a', {'class' : 'linklist__link'})
        for link in links:
            full_link = 'https:' + link.get('href')
            title = link.text
            files_dic.update({ title : full_link })
        return files_dic
    except:
        pass
```

```python
def get_specs_pdf(soup):
    try:
        base_link = 'https://www.grohe-mena.com'
        pdf_link = soup.find('a', {'class' : 'product-box__button link link--downloadBadgePrimary'}).get('href')
        pdf = base_link + pdf_link
        return pdf
    except:
        pass
```

```python
def get_prod_descr(soup):
    try:
        descr = soup.find('ul', {'class' : 'columns columns-3 list'}).findAll('li')
        s = str(descr[0].text)

        for i in descr:
            i = i.find_next()
            s = s + ', ' + str(i.text)
        return s
    except:
        pass
```

```python
def get_spare_parts_img(soup):
    try:
        img = soup.find('img', {'class' : 'img-fluid'}).get('src')
        img = 'https:' + img
        return img
    except:
        pass
```
